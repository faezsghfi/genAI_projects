```python
import torch
from diffusers import StableDiffusionPipeline

# لود مدل (نیاز به GPU داره برای عملکرد بهتر)
pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    torch_dtype=torch.float16
).to("cuda")
```

```python
prompt = "a futuristic city with flying cars at sunset"
image = pipe(prompt).images[0]
image.show()
```

```python
import gradio as gr

def generate_image(prompt):
    image = pipe(prompt).images[0]
    return image

gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(lines=2, placeholder="مثلاً: A cyberpunk cat with glasses"),
    outputs="image",
    title="🧠 Text to Image Generator",
    description="با وارد کردن متن، تصویر بساز!",
    theme="default"
).launch()
```
